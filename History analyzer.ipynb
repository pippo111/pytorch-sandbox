```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
histories = {}
epochs = np.arange(50)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
```

```python
history_path = '/home/filip/Projekty/ML/histories/wmh_Unet_Adam_histories'

history_files = {
    'bce': 'wmh_Unet_Adam_bce_bs-16_f-16_history.npy',
    #'weighted_bce': 'wmh_Unet_Adam_weighted_bce_bs-16_f-16_history.npy',
    'dice': 'wmh_Unet_Adam_dice_bs-16_f-16_history.npy',
    #'gdl': 'wmh_Unet_Adam_gdl_bs-16_f-16_history.npy',
    #'surface': 'wmh_Unet_Adam_surface_bs-16_f-16_history.npy',
    #'boundary_bce': 'wmh_Unet_Adam_boundary_bce_bs-16_f-16_history.npy',
    'boundary_dice': 'wmh_Unet_Adam_boundary_dice_bs-16_f-16_history.npy',
    #'boundary_wbce': 'wmh_Unet_Adam_boundary_wbce_bs-16_f-16_history.npy'
}
```

```python
for file in history_files:
    full_path = os.path.join(history_path, history_files[file])
    histories[file] = np.load(full_path, allow_pickle=True)
```

```python
metric = 'f_total'
```

```python
fig = plt.figure(figsize=(20,10))

for i, loss_name in enumerate(histories):
    for test in np.atleast_1d(histories[loss_name]):
        
        plt.plot(np.arange(len(test[metric])), test[metric], color=colors[i], label=loss_name)

plt.yscale('symlog')
plt.legend(loc=1, prop={'size': 18})
plt.title(metric, fontsize=20)
plt.show()
```
